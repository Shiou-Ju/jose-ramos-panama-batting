```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Horizontal_Ending': ['outside', 'middle', 'inside', 'outside', 'middle', 'inside', 'outside', 'middle', 'inside'],
    'Vertical_Ending': ['high', 'high', 'high',  'middle', 'middle', 'middle', 'low', 'low', 'low',],
    # 假設隨機，先看看效果
    'Swing_Intention_Rate': np.random.rand(9),
}
df = pd.DataFrame(data)

pivot_table = df.pivot(index='Vertical_Ending',
                       columns='Horizontal_Ending', values='Swing_Intention_Rate')

# plt.figure(figsize=(10, 8))
# sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="coolwarm", square=True)
# plt.title("Swing Intention Rate by Pitch Ending Zone")
# plt.show()

expected_vertical_order = ['high', 'middle', 'low']
expected_horizontal_order = ['outside', 'middle', 'inside']

pivot_table_reordered = pivot_table.reindex(index=expected_vertical_order,
                                            columns=expected_horizontal_order)

plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table_reordered, annot=True,
            fmt=".2f", cmap="coolwarm", square=True)
plt.title("Swing Intention Rate")
plt.show()
```
